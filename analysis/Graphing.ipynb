```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
def plotGraph(file):
    df = pd.read_csv(file)
    df.columns = ['Position','Frequency']
    df.ylabel = 'Frequency'
    df.xlabel = 'Position'
    df.plot(x='Position', y='Frequency', kind='scatter', title=file.split('/')[3].split('.')[0])
```

```python
plotGraph('../data/cleanedData/ARID1A.csv')
plotGraph('../data/cleanedData/ARID1B.csv')
plotGraph('../data/cleanedData/ARID2.csv')
```
